```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
plt.bar(kategori,data)
plt.grid()
plt.show()
```

```python
data = [25,45,55,125,225]
kategori = ['A','B','C','D','E']
plt.barh(kategori,data)
plt.grid()
plt.show()
```
