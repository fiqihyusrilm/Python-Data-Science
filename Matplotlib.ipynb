```python
import numpy as np
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Judul Grafik')
plt.show()
```

```python
plt.subplot(211)
plt.plot(x,y, color='red')

plt.subplot(212)
plt.plot(x,y,color='blue')
```

```python
plt.subplot(121)
plt.plot(x,y, color='red')

plt.subplot(122)
plt.plot(x,y,color='blue')
```

```python
x = np.arange(1,20,0.5)
y = x**2 
```

```python
# OOP matplotlib
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul Grafik')
plt.show()
```
